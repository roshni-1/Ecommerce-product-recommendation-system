## 4. Recommendation System Design:
The recommendation system utilized collaborative filtering techniques. A user-item matrix was created from the interactions data. Missing values were handled appropriately, ensuring data integrity. Matrix factorization using Truncated Singular Value Decomposition (SVD) was applied to reduce dimensionality and capture latent features. Cosine similarity was calculated to identify similar products for recommendations.

```python
# Remove duplicate entries keeping the last occurrence
interactions = interactions.drop_duplicates(subset=['user_id', 'product_id'], keep='last')

# Create user-item matrix
user_item_matrix = interactions.pivot(index='user_id', columns='product_id', values='rating').fillna(0)

# Perform matrix factorization using SVD
svd = TruncatedSVD(n_components=min(len(user_item_matrix), 50), random_state=42)
user_item_matrix_svd = svd.fit_transform(user_item_matrix)

```

```python
# Function to get recommendations for a user
def get_recommendations(user_id, num_recommendations=5):
    if user_id in user_item_matrix.index:
        user_ratings = user_item_matrix.loc[user_id].values.reshape(1, -1)
        cosine_similarities = cosine_similarity(user_ratings, user_item_matrix.values)
        product_indices = cosine_similarities.argsort()[0][::-1][:num_recommendations]
        recommended_product_ids = user_item_matrix.columns[product_indices].tolist()
        return recommended_product_ids
    else:
        return f"User with ID '{user_id}' not found in the dataset."

# Example usage
user_id = 136  # Replace with a user ID from your dataset
recommendations = get_recommendations(user_id, num_recommendations=5)
print('Recommended Products:', recommendations)
```
